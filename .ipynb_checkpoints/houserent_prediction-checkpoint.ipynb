```python
# Define categorical and numerical columns
categorical_cols = ["seller_type", "layout_type", "property_type", "furnish_type", "city"]
numerical_cols = ["bedroom", "area", "bathroom"]
```

```python

# One-Hot Encode categorical features
encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
encoded_features = encoder.fit_transform(df[categorical_cols])
```

```python
# Convert encoded features into a DataFrame
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out())
```

```python
# Reset index to avoid misalignment issues
df_numeric = df[numerical_cols].reset_index(drop=True)
encoded_df = encoded_df.reset_index(drop=True)
```

```python
encoded_df
```

```python
df2 = pd.concat([df_numeric, encoded_df,df["price"]], axis=1) 
```
